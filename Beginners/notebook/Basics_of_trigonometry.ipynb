### $\text{If: } sin(x) + cos(x) = \frac{1}{5}\text{, find the value of: } tan(x)$

$\text{Squaring both sides: } {(sin(x) + cos(x))}^2 \implies sin^2(x)+cos^2(x)+2{\cdot}sin(x){\cdot}\cos(x)=\frac{1}{25}$

$\because {sin^2(x)+cos^2(x)=1} \implies 1+2{\cdot}sin(x){\cdot}\cos(x)=\frac{1}{25}$

$\therefore sin(x){\cdot}\cos(x)=-\frac{12}{25}$

$\because {cos(x)=\frac{1}{5}-sin(x)} \implies sin(x){\cdot}({\frac{1}{5}-sin(x)})=-\frac{12}{25}$

$\therefore 5{\cdot}sin^2(x)-sin(x)=\frac{12}{5}$

$\therefore 25{\cdot}sin^2(x)-5{\cdot}sin(x)-12=0$

$\text{Applying the solution to a quadratic euqation of the form: }ax^2+bx+c=0\text{, }x = \frac{-b\pm\sqrt{b^2-4{\cdot}a{\cdot}c}}{2{\cdot}a}$

$\therefore sin(x)=\frac{5\pm\sqrt{5^2+(4{\cdot}25{\cdot}12})}{2{\cdot}25}$

$\therefore sin(x)=\frac{1\pm{7}}{10}$

$\therefore sin(x)=\begin{cases} \frac{4}{5} \\
                     -\frac{3}{5}
       \end{cases}$

$\because {cos(x)=\frac{1}{5}-sin(x)} \implies cos(x)=\begin{cases} -\frac{3}{5} & sin(x) = \frac{4}{5} \\
\\
                     \frac{4}{5} & sin(x) = -\frac{3}{5}
       \end{cases}$

$\because {tan(x)=\frac{sin(x)}{cos(x)}} \implies tan(x)=\begin{cases} -\frac{4}{3} & sin(x) = \frac{4}{5}, cos(x)=-\frac{3}{5} \\
\\
                     -\frac{3}{4} & sin(x) = -\frac{3}{5}, cos(x) = \frac{4}{5}
       \end{cases}$

```python
import math, numpy as np
```

$\text{Evaluating: } sin(sin^{-1}(\frac{4}{5})) + cos(cos^{-1}(-\frac{3}{5})) \approx sin(sin^{-1}(-\frac{3}{5})) + cos(cos^{-1}(\frac{4}{5})) \approx \frac{1}{5}$

```python
print(

    np.isclose(

        np.sin(np.arcsin(4 / 5))  + np.cos(np.arccos(-3 / 5)),

        np.sin(np.arcsin(-3 / 5)) + np.cos(np.arccos(4 / 5)),

        1 / 5

    )

)
```

$\text{Evaluating: } cos^{-1}(-\frac{3}{5}) \text{ and } cos^{-1}(\frac{4}{5})$

```python
print(

    np.arccos(-3 / 5), 

    np.arccos(4 / 5)

)
```

$\text{Evaluating: } tan^{-1}(-\frac{4}{3}) \text{ and } tan^{-1}(-\frac{3}{4})$
